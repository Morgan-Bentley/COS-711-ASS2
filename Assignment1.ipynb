# Part 1: Data preprocessing step

```python
import pandas as pd
```

```python
# Import Almond.csv data for preprocessing
almond = pd.read_csv('Almond.csv')
# view first ten rows of the data
almond.head(17)
```

```python
print('Distribution of the types of Almond in the data')
print('MAMRA:',almond['Type'].value_counts()['MAMRA'])
print('REGULAR:', almond['Type'].value_counts()['REGULAR'])
print('SANORA:', almond['Type'].value_counts()['SANORA'])
```

```python
# drop the first column
almond = almond.drop('Unnamed: 0', axis=1)
# show the shape of the data
almond
```

```python
# check for duplicate rows
almond.duplicated().sum()
```

```python
# drop duplicate rows
almond = almond.drop_duplicates()
almond.duplicated().sum()
```

```python
# show number of null values in each column
almond.isnull().sum()
```

```python
# correlation matrix for numerical columns
# use heatmap to visualize the correlation matrix, specify size of plot
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
sns.heatmap(almond.iloc[:, :-1].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#iterative imputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer(random_state=45, verbose=2)

# columns to impute
columns = ['Length (major axis)',
           'Width (minor axis)',
           'Thickness (depth)',
           'Area',
           'Perimeter',
           'Convex hull(convex area)'
          ] 

almond_imputed = almond.copy()
# drop last column from almond_imputed
almond_imputed = almond_imputed.iloc[:, :-1]

almond_imputed[columns] = imputer.fit_transform(almond_imputed[columns])
almond_imputed = pd.DataFrame(almond_imputed, columns=almond.columns)
almond_imputed['Type'] = almond['Type']

# show number of null values in each column
almond_imputed.isnull().sum()
```

```python
# print the first few rows of the data
almond_imputed
```

```python
import numpy as np
# use the imputed data to impute the missing values in the columns from derived features
almond_imputed['Roundness'] = 4 * almond_imputed['Area'] / (np.pi * almond_imputed['Length (major axis)'] ** 2)
almond_imputed['Aspect Ratio'] = almond_imputed['Length (major axis)'] / almond_imputed['Width (minor axis)']
almond_imputed['Eccentricity'] = np.sqrt(1 - (almond_imputed['Width (minor axis)'] / almond_imputed['Length (major axis)']) ** 2)
```

```python
# show number of remaining null values in each column
almond_imputed.isnull().sum()
```

```python
almond_imputed = almond_imputed.drop('Eccentricity', axis=1)
almond_imputed = almond_imputed.drop('Aspect Ratio', axis=1)
# show the data
almond_imputed.isnull().sum()
```

```python
# convert Type to numerical values, MAMRA=0, REGULAR=1, SANORA=2
almond_imputed['Type'] = almond_imputed['Type'].map({'MAMRA': 0, 'REGULAR': 1, 'SANORA': 2})
almond_imputed.dtypes
```

```python
# shuffle the data and then split the data into Training, Validation and Test set.
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler

# shuffle the data
almond_imputed = almond_imputed.sample(frac=1, random_state=45).reset_index(drop=True)

x = almond_imputed.drop('Type', axis=1)
y = almond_imputed['Type']

# split the data into training and test set
x_train_val, x_test, y_train_val, y_test = train_test_split(x, y, test_size=0.2, random_state=45)

scaler = StandardScaler()
x_train_val = scaler.fit_transform(x_train_val)
x_test = scaler.transform(x_test)

# split the training set into training and validation set
x_train, x_val, y_train, y_val = train_test_split(x_train_val, y_train_val, test_size=0.2, random_state=45)
```

```python
# convert the data to tensor
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# convert the data to tensor, and then to DataLoader using a method
def create_data_loader(x, y, batch_size, shuffle, drop_last=True):
    x_tensor = torch.tensor(x, dtype=torch.float32)
    y_tensor = torch.tensor(y.values, dtype=torch.long)
    dataset = TensorDataset(x_tensor, y_tensor)
    return DataLoader(dataset, batch_size=batch_size, shuffle=shuffle, drop_last=drop_last)
```

# Part 2: building and training Adam and Rprop optimizer

```python
# build a dynamic neural network model
class AlmondClassifier(nn.Module):
    def __init__(self, input_features, hidden_layers, output_features, activation_function):
        super(AlmondClassifier, self).__init__()
        layers = []
        prev_layer = input_features

        for layer in hidden_layers:
            layers.append(nn.Linear(prev_layer, layer))
            layers.append(activation_function)
            prev_layer = layer

        layers.append(nn.Linear(prev_layer, output_features))
        self.model = nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x)
```

```python
# use optuna optimization to find the best hyperparameters for the model
import optuna
from optuna import Trial, visualization
from optuna.samplers import TPESampler
from optuna.visualization.matplotlib import plot_optimization_history, plot_param_importances, plot_slice
from optuna.pruners import MedianPruner

# define the objective function

def objective(trial: Trial, optimizer_name='Rprop'):
    # define the hyperparameters to optimize
    input_features = x_train.shape[1]
    output_features = 3
    number_hidden_layers = trial.suggest_int('number_hidden_layers', 1, 4)
    hidden_layers = [trial.suggest_int(f'hidden_layer_{i}', 12, 256) for i in range(number_hidden_layers)]
    activation_function_name = trial.suggest_categorical('activation_function', ['ReLU', 'Sigmoid', 'Tanh'])
    activation_function = getattr(nn, activation_function_name)()
    number_epochs = trial.suggest_int('number_epochs', 50, 220)
    batch_size = trial.suggest_categorical('batch_size', [16, 32, 64, 128, 256])

    # create the model
    model = AlmondClassifier(input_features, hidden_layers, output_features, activation_function)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()

    if optimizer_name == 'Adam':
        learning_rate = trial.suggest_float('learning_rate', 1e-5, 1e-1, log=True)
        optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    elif optimizer_name == 'Rprop':
        optimizer = optim.Rprop(model.parameters())

    # create the data loaders
    train_loader = create_data_loader(x_train, y_train, batch_size, shuffle=True)
    val_loader = create_data_loader(x_val, y_val, batch_size, shuffle=False)

    

    # train the model
    for epoch in range(number_epochs):
        model.train()
        for batch_input, batch_output in train_loader:
            batch_input = batch_input.to(device)
            batch_output = batch_output.to(device)
            output = model(batch_input)
            loss = criterion(output, batch_output)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        model.eval()
        val_loss = 0
        correct = 0
        total = 0
        with torch.no_grad():
            for batch_input, batch_output in val_loader:
                batch_input = batch_input.to(device)
                batch_output = batch_output.to(device)
                output = model(batch_input)
                loss = criterion(output, batch_output)
                val_loss += loss.item()
                _, predicted = torch.max(output, 1)
                total += batch_output.size(0)
                correct += (predicted == batch_output).sum().item()

        accuracy = correct / total
        val_loss /= len(val_loader)
        trial.report(accuracy, epoch)
        if trial.should_prune():
            raise optuna.TrialPruned()
        
    return accuracy

# learning for Adam optimizer
sampler = TPESampler(seed=48)
adam_study = optuna.create_study(study_name='Almond Classifier', direction="maximize", sampler=sampler, pruner=MedianPruner())
adam_study.optimize(lambda trial: objective(trial, 'Adam'), n_trials=200)

# learning for Rprop optimizer
sampler = TPESampler(seed=48)
rprop_study = optuna.create_study(study_name='Almond Classifier', direction="maximize", sampler=sampler, pruner=MedianPruner())
rprop_study.optimize(lambda trial: objective(trial, 'Rprop'), n_trials=200)

# plot the study for Adam optimizer
plot_param_importances(adam_study)
plot_optimization_history(adam_study)
plot_slice(adam_study)

# plot the study for Rprop optimizer
plot_param_importances(rprop_study)
plot_optimization_history(rprop_study)
plot_slice(rprop_study)

# get the best hyperparameters for both optimizers
adam_best_params = adam_study.best_params
rprop_best_params = rprop_study.best_params

print('Best hyperparameters for Adam optimizer:', adam_best_params)
print('Best hyperparameters for Rprop optimizer:', rprop_best_params)

    
```

```python
# train the model using the best hyperparameters for Adam optimizer and Rprop optimizer
def trainer( best_params, optimizer_name, patience=10):
    input_features = x_train.shape[1]
    output_features = 3
    hidden_layers = [best_params[f'hidden_layer_{i}'] for i in range(best_params['number_hidden_layers'])]
    activation_function = nn.ReLU()
    number_epochs = best_params['number_epochs']
    batch_size = best_params['batch_size']

    model = AlmondClassifier(input_features, hidden_layers, output_features, activation_function)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    if optimizer_name == 'Adam':
        learning_rate = best_params['learning_rate']
        optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    elif optimizer_name == 'Rprop':
        optimizer = optim.Rprop(model.parameters())
        


    train_loader = create_data_loader(x_train_val, y_train_val, batch_size, shuffle=True, drop_last=False)
    val_loader = create_data_loader(x_val, y_val, batch_size, shuffle=False)

    

    train_losses = []
    train_accuracies = []
    val_losses = []
    val_accuracies = []
    for epoch in range(number_epochs):
        train_loss = 0
        correct = 0
        total = 0
        model.train()
        for batch_input, batch_output in train_loader:
            batch_input = batch_input.to(device)
            batch_output = batch_output.to(device)
            output = model(batch_input)
            loss = criterion(output, batch_output)
            train_loss += loss.item()
            _, predicted = torch.max(output, 1)
            total += batch_output.size(0)
            correct += (predicted == batch_output).sum().item()


            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        
        train_loss /= len(train_loader)
        train_accuracy = correct / total
        train_losses.append(train_loss)
        train_accuracies.append(train_accuracy)

        model.eval()
        val_loss = 0
        correct = 0
        total = 0
        with torch.no_grad():
            for batch_input, batch_output in val_loader:
                batch_input = batch_input.to(device)
                batch_output = batch_output.to(device)
                output = model(batch_input)
                loss = criterion(output, batch_output)
                val_loss += loss.item()
                _, predicted = torch.max(output, 1)
                total += batch_output.size(0)
                correct += (predicted == batch_output).sum().item()

        accuracy = correct / total
        val_loss /= len(val_loader)
        val_losses.append(val_loss)
        val_accuracies.append(accuracy)
        print(f'Epoch: {epoch}, Validation Loss: {val_loss}, Validation Accuracy: {accuracy}')
        # add early stopping here, with patience of 10
        best_loss = float('inf')
        no_improvement = 0
        if epoch > patience and val_loss < best_loss:
            best_loss = val_loss
            best_model = model.state_dict()
            no_improvement = 0
        elif epoch > patience and val_loss > best_loss:
            no_improvement += 1
            if no_improvement > patience:
                model.load_state_dict(best_model)
                print('Early stopping at epoch:', epoch)
                break

    return train_losses, train_accuracies, val_losses, val_accuracies, model

#adam_best_params
adam_train_losses, adam_train_accuracies, adam_val_losses, adam_val_accuracies, adam_model = trainer(adam_best_params, 'Adam')
#rprop_best_params
rprop_train_losses, rprop_train_accuracies, rprop_val_losses, rprop_val_accuracies, rprop_model = trainer(rprop_best_params, 'Rprop')

```

```python
# test the rprop model and adam model on the test set (accuracy, confusion matrix, classification report)
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import precision_score, recall_score, f1_score
def test_model(model, loader):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    model.eval()
    predictions = []
    targets = []
    correct = 0
    total = 0
    with torch.no_grad():
        for batch_input, batch_output in loader:
            batch_input = batch_input.to(device)
            batch_output = batch_output.to(device)
            output = model(batch_input)
            _, predicted = torch.max(output, 1)
            total += batch_output.size(0)
            correct += (predicted == batch_output).sum().item()
            predictions.extend(predicted.cpu().numpy())
            targets.extend(batch_output.cpu().numpy())

    accuracy = correct / total

    precision = precision_score(targets, predictions, average='weighted')
    recall = recall_score(targets, predictions, average='weighted')
    f1 = f1_score(targets, predictions, average='weighted')
    return [accuracy , precision, recall, f1], confusion_matrix(targets, predictions)

adam_test_loader = create_data_loader(x_test, y_test, adam_best_params['batch_size'], shuffle=False)
rprop_test_loader = create_data_loader(x_test, y_test, rprop_best_params['batch_size'], shuffle=False)

adam_test_metrics, adam_confusion_matrix = test_model(adam_model, adam_test_loader)
rprop_test_metrics, rprop_confusion_matrix = test_model(rprop_model, rprop_test_loader)
```

```python
# plot confusion matrix
def plot_confusion_matrix(confusion_matrix, classes, title):
    plt.figure(figsize=(10, 10))
    sns.heatmap(confusion_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=classes, yticklabels=classes)
    plt.title(title)
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.show()

# plot confusion matrix for adam model
plot_confusion_matrix(adam_confusion_matrix, ['MAMRA', 'REGULAR', 'SANORA'], 'Adam Confusion Matrix')
# plot confusion matrix for rprop model
plot_confusion_matrix(rprop_confusion_matrix, ['MAMRA', 'REGULAR', 'SANORA'], 'Rprop Confusion Matrix')
```

```python
# tabulate the metrics for both optimizers
def tabulate_metrics(metrics, title):
    data = {'Metrics': ['Accuracy', 'Precision', 'Recall', 'F1 Score'], 'Values': metrics}
    df = pd.DataFrame(data)
    print(title)
    print(df)

tabulate_metrics(adam_test_metrics, 'Adam Test Metrics')
tabulate_metrics(rprop_test_metrics, 'Rprop Test Metrics')
```
